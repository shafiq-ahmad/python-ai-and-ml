```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
df = sns.load_dataset('tips')
df.head(5)
```

```python
df['day'].value_counts()
```

```python
# let's encode the time in labelencoder with sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df.head()
```

```python
# ordinal encoding the day column using specific order
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
# one hot encoding on day column
ohe = OneHotEncoder()
ohe.fit_transform(df[['day']]).toarray()
```

## Data Discretization

```python
# import KBinsDescretizer
from sklearn.preprocessing import KBinsDiscretizer
```

```python
# load titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.histplot(titanic['age'], kde=True)
```

```python
# impute missing values
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['fare'] = titanic['fare'].fillna(titanic['fare'].median())
```

```python
# age discretization
age_discretizer = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='uniform')
titanic['age_bin'] = age_discretizer.fit_transform(titanic[['age']])
df.head()
```
